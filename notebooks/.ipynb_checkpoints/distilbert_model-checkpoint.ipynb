# Toxicity Detection - DistilBERT Fine-tuning

### DistilBERT Fine-tuning for Toxicity Detection 
### **Goal:** Fine-tune DistilBERT for multi-label toxicity classification

### **Expected Improvements over Baseline:**
### - Better contextual understanding
### - Handles obfuscated text (bull$hit → bullshit)
### - Improved performance on rare labels
### - Higher F1 scores overall

```python
# Imports
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW  # Changed: Import from torch instead of transformers
from transformers import (
    DistilBertTokenizer, 
    DistilBertForSequenceClassification,
    get_linear_schedule_with_warmup
)
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, multilabel_confusion_matrix
)
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm
import warnings

print("Imports complete")
```

### 1. Setup and Configuration

```python
device = torch.device('cuda' if torch.cuda.is_available() 
                     else 'mps' if torch.backends.mps.is_available() 
                     else 'cpu')

print(f"Using device: {device}")
if device.type == 'cuda':
    print(f"GPU: {torch.cuda.get_device_name(0)}")
elif device.type == 'mps':
    print("Apple Silicon GPU (MPS)")
else:
    print("CPU - Training will be slower but still works!")
```

```python
# Configuration
CONFIG = {
    'model_name': 'distilbert-base-uncased',
    'max_length': 128,        # Token length limit
    'batch_size': 16,         # Reduce to 8 if out of memory
    'num_epochs': 3,          # 3-5 epochs is typical
    'learning_rate': 2e-5,
    'warmup_steps': 500,
    'weight_decay': 0.01,
    'random_seed': 42
}

print("\nConfiguration:")
for key, value in CONFIG.items():
    print(f"  {key}: {value}")

# Set random seed
torch.manual_seed(CONFIG['random_seed'])
np.random.seed(CONFIG['random_seed'])
```

### 2. Load Data


```python
print("Loading data...")

# Load preprocessed data from Day 1
train_df = pd.read_csv('../data/train.csv')
label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

# Use same split as baseline
from sklearn.model_selection import train_test_split

train_data, val_data = train_test_split(
    train_df,
    test_size=0.2,
    stratify=train_df['toxic'],
    random_state=42
)

print(f"\nTrain set: {len(train_data):,} samples")
print(f"Val set:   {len(val_data):,} samples")

# Check label distribution
print("\nLabel distribution (train set):")
for col in label_cols:
    count = train_data[col].sum()
    pct = count / len(train_data) * 100
    print(f"  {col:15s}: {count:6,} ({pct:5.2f}%)")
```

### 3. Tokenization

```python
print("Loading tokenizer...")
tokenizer = DistilBertTokenizer.from_pretrained(CONFIG['model_name'])

print(f"✓ Tokenizer loaded: {CONFIG['model_name']}")
print(f"  Vocab size: {tokenizer.vocab_size:,}")
print(f"  Max length: {CONFIG['max_length']}")

# %%
# Test tokenization
sample_text = train_data.iloc[0]['comment_text']
print("\nSample tokenization:")
print(f"Original: {sample_text[:100]}...")

encoded = tokenizer.encode_plus(
    sample_text,
    max_length=CONFIG['max_length'],
    padding='max_length',
    truncation=True,
    return_tensors='pt'
)

print(f"\nTokenized shape: {encoded['input_ids'].shape}")
print(f"First 20 tokens: {encoded['input_ids'][0][:20].tolist()}")
print(f"Decoded: {tokenizer.decode(encoded['input_ids'][0][:20])}")
```

### 4. Create PyTorch Dataset

```python
class ToxicityDataset(Dataset):
    """Dataset for toxicity classification"""
    
    def __init__(self, dataframe, tokenizer, max_length, label_cols):
        self.data = dataframe.reset_index(drop=True)
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.label_cols = label_cols
        
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        # Get text and labels
        text = str(self.data.loc[idx, 'comment_text'])
        labels = self.data.loc[idx, self.label_cols].values.astype(float)
        
        # Tokenize
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.FloatTensor(labels)
        }
# Create datasets
print("Creating datasets...")

train_dataset = ToxicityDataset(
    train_data, 
    tokenizer, 
    CONFIG['max_length'],
    label_cols
)

val_dataset = ToxicityDataset(
    val_data,
    tokenizer,
    CONFIG['max_length'],
    label_cols
)

print(f"✓ Train dataset: {len(train_dataset):,} samples")
print(f"✓ Val dataset:   {len(val_dataset):,} samples")

# Test dataset
sample = train_dataset[0]
print(f"\nSample batch shapes:")
print(f"  input_ids: {sample['input_ids'].shape}")
print(f"  attention_mask: {sample['attention_mask'].shape}")
print(f"  labels: {sample['labels'].shape}")
```

### 5. Create DataLoaders

```python
train_loader = DataLoader(
    train_dataset,
    batch_size=CONFIG['batch_size'],
    shuffle=True,
    num_workers=0  # Set to 0 for compatibility
)

val_loader = DataLoader(
    val_dataset,
    batch_size=CONFIG['batch_size'],
    shuffle=False,
    num_workers=0
)

print(f"Train batches: {len(train_loader):,}")
print(f"Val batches:   {len(val_loader):,}")
print(f"\nTotal training steps: {len(train_loader) * CONFIG['num_epochs']:,}")
```

### 6. Load Model

```python
print("Loading DistilBERT model...")

model = DistilBertForSequenceClassification.from_pretrained(
    CONFIG['model_name'],
    num_labels=len(label_cols),
    problem_type="multi_label_classification"
)

model = model.to(device)

print(f"✓ Model loaded and moved to {device}")
print(f"  Parameters: {sum(p.numel() for p in model.parameters()):,}")
print(f"  Trainable parameters: {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")
```

### 7. Setup Training

```python
# Optimizer
optimizer = AdamW(
    model.parameters(),
    lr=CONFIG['learning_rate'],
    weight_decay=CONFIG['weight_decay']
)

# Learning rate scheduler
total_steps = len(train_loader) * CONFIG['num_epochs']
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=CONFIG['warmup_steps'],
    num_training_steps=total_steps
)

print(" Optimizer and scheduler configured")
print(f"  Learning rate: {CONFIG['learning_rate']}")
print(f"  Warmup steps: {CONFIG['warmup_steps']}")
print(f"  Total steps: {total_steps:,}")
```

### 8. Training Loop

```python
def train_epoch(model, data_loader, optimizer, scheduler, device):
    """Train for one epoch"""
    model.train()
    losses = []
    
    progress_bar = tqdm(data_loader, desc="Training")
    
    for batch in progress_bar:
        # Move to device
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)
        
        # Forward pass
        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels
        )
        
        loss = outputs.loss
        losses.append(loss.item())
        
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        
        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        optimizer.step()
        scheduler.step()
        
        # Update progress bar
        progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})
    
    return np.mean(losses)

# %%
def evaluate(model, data_loader, device):
    """Evaluate model"""
    model.eval()
    losses = []
    all_preds = []
    all_labels = []
    
    progress_bar = tqdm(data_loader, desc="Evaluating")
    
    with torch.no_grad():
        for batch in progress_bar:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)
            
            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=labels
            )
            
            loss = outputs.loss
            losses.append(loss.item())
            
            # Get predictions (sigmoid for multi-label)
            logits = outputs.logits
            probs = torch.sigmoid(logits)
            preds = (probs > 0.5).float()
            
            all_preds.append(preds.cpu().numpy())
            all_labels.append(labels.cpu().numpy())
    
    # Concatenate all batches
    all_preds = np.vstack(all_preds)
    all_labels = np.vstack(all_labels)
    
    return np.mean(losses), all_preds, all_labels
```

### 9. Train Model

```python
print("\n" + "="*60)
print("TRAINING DISTILBERT")
print("="*60)
print(f"\nThis will take approximately {CONFIG['num_epochs'] * len(train_loader) * 2 / 60:.0f}-{CONFIG['num_epochs'] * len(train_loader) * 3 / 60:.0f} minutes on GPU")
print("You can take a break - progress will be shown!\n")

history = {
    'train_loss': [],
    'val_loss': [],
    'val_f1': []
}

best_f1 = 0
patience_counter = 0
patience = 2  # Early stopping patience

for epoch in range(CONFIG['num_epochs']):
    print(f"\nEpoch {epoch + 1}/{CONFIG['num_epochs']}")
    print("-" * 60)
    
    # Train
    train_loss = train_epoch(model, train_loader, optimizer, scheduler, device)
    print(f"Train Loss: {train_loss:.4f}")
    
    # Evaluate
    val_loss, val_preds, val_labels = evaluate(model, val_loader, device)
    
    # Calculate metrics
    macro_f1 = f1_score(val_labels, val_preds, average='macro')
    micro_f1 = f1_score(val_labels, val_preds, average='micro')
    
    print(f"Val Loss:   {val_loss:.4f}")
    print(f"Macro F1:   {macro_f1:.4f}")
    print(f"Micro F1:   {micro_f1:.4f}")
    
    # Save history
    history['train_loss'].append(train_loss)
    history['val_loss'].append(val_loss)
    history['val_f1'].append(macro_f1)
    
    # Save best model
    if macro_f1 > best_f1:
        best_f1 = macro_f1
        patience_counter = 0
        print(f"✓ New best F1: {best_f1:.4f} - Saving model...")
        
        import os
        os.makedirs('../models/distilbert', exist_ok=True)
        model.save_pretrained('../models/distilbert')
        tokenizer.save_pretrained('../models/distilbert')
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping triggered (patience={patience})")
            break

print("\n" + "="*60)
print("TRAINING COMPLETE!")
print("="*60)
print(f"Best Macro F1: {best_f1:.4f}")
```

### 10. Training History Visualization

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loss plot
axes[0].plot(history['train_loss'], label='Train Loss', marker='o')
axes[0].plot(history['val_loss'], label='Val Loss', marker='s')
axes[0].set_xlabel('Epoch', fontsize=12)
axes[0].set_ylabel('Loss', fontsize=12)
axes[0].set_title('Training and Validation Loss', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(alpha=0.3)

# F1 plot
axes[1].plot(history['val_f1'], label='Val Macro F1', marker='o', color='green')
axes[1].axhline(y=best_f1, color='red', linestyle='--', label=f'Best F1: {best_f1:.4f}')
axes[1].set_xlabel('Epoch', fontsize=12)
axes[1].set_ylabel('Macro F1', fontsize=12)
axes[1].set_title('Validation F1 Score', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('../results/distilbert_training_history.png', dpi=300, bbox_inches='tight')
plt.show()

print("Saved: results/distilbert_training_history.png")
```

### 11. Load Best Model and Evaluate

```python
print("Loading best model...")
model = DistilBertForSequenceClassification.from_pretrained('../models/distilbert')
model = model.to(device)
print("✓ Best model loaded")

# Get final predictions
print("\nEvaluating on validation set...")
val_loss, val_preds, val_labels = evaluate(model, val_loader, device)
```

### 12. Detailed Metrics

```python
print("\n" + "="*60)
print("DISTILBERT EVALUATION RESULTS")
print("="*60)

# Overall metrics
print("\n--- OVERALL METRICS ---")
hamming_loss = 1 - accuracy_score(val_labels, val_preds)
exact_match = accuracy_score(val_labels, val_preds)
macro_f1 = f1_score(val_labels, val_preds, average='macro')
micro_f1 = f1_score(val_labels, val_preds, average='micro')

print(f"Hamming Loss:      {hamming_loss:.4f}")
print(f"Exact Match Ratio: {exact_match:.4f}")
print(f"Macro F1:          {macro_f1:.4f}")
print(f"Micro F1:          {micro_f1:.4f}")


# Per-class metrics
print("\n--- PER-CLASS METRICS ---")
print(f"{'Label':<15} {'Precision':>10} {'Recall':>10} {'F1-Score':>10} {'Support':>10}")
print("-" * 65)

distilbert_metrics = []

for i, label in enumerate(label_cols):
    precision = precision_score(val_labels[:, i], val_preds[:, i], zero_division=0)
    recall = recall_score(val_labels[:, i], val_preds[:, i], zero_division=0)
    f1 = f1_score(val_labels[:, i], val_preds[:, i], zero_division=0)
    support = int(val_labels[:, i].sum())
    
    print(f"{label:<15} {precision:>10.3f} {recall:>10.3f} {f1:>10.3f} {support:>10,}")
    
    distilbert_metrics.append({
        'Label': label,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Support': support
    })

distilbert_metrics_df = pd.DataFrame(distilbert_metrics)

# Save metrics
import os
os.makedirs('../results', exist_ok=True)
distilbert_metrics_df.to_csv('../results/distilbert_metrics.csv', index=False)
print("\n Saved: results/distilbert_metrics.csv")
```

### 13. Compare with Baseline

```python
print("\n" + "="*60)
print("BASELINE vs DISTILBERT COMPARISON")
print("="*60)

# Load baseline metrics
baseline_metrics_df = pd.read_csv('../results/baseline_metrics.csv')

# Merge for comparison
comparison_df = baseline_metrics_df[['Label', 'F1-Score']].rename(columns={'F1-Score': 'Baseline_F1'})
comparison_df = comparison_df.merge(
    distilbert_metrics_df[['Label', 'F1-Score']].rename(columns={'F1-Score': 'DistilBERT_F1'}),
    on='Label'
)
comparison_df['Improvement'] = comparison_df['DistilBERT_F1'] - comparison_df['Baseline_F1']
comparison_df['Improvement_Pct'] = (comparison_df['Improvement'] / comparison_df['Baseline_F1'] * 100)

print("\n" + f"{'Label':<15} {'Baseline':>10} {'DistilBERT':>12} {'Improvement':>12} {'% Change':>10}")
print("-" * 65)

for idx, row in comparison_df.iterrows():
    print(f"{row['Label']:<15} {row['Baseline_F1']:>10.3f} {row['DistilBERT_F1']:>12.3f} "
          f"{row['Improvement']:>12.3f} {row['Improvement_Pct']:>9.1f}%")

# Overall comparison
baseline_macro = baseline_metrics_df['F1-Score'].mean()
distilbert_macro = distilbert_metrics_df['F1-Score'].mean()
improvement = distilbert_macro - baseline_macro

print("\n" + "-" * 65)
print(f"{'OVERALL':<15} {baseline_macro:>10.3f} {distilbert_macro:>12.3f} "
      f"{improvement:>12.3f} {improvement/baseline_macro*100:>9.1f}%")

# Save comparison
comparison_df.to_csv('../results/model_comparison.csv', index=False)
print("\n Saved: results/model_comparison.csv")
```

### 15. Sample Predictions

```python
print("\n" + "="*60)
print("SAMPLE PREDICTIONS - Error Cases from Baseline")
print("="*60)

# Load baseline predictions to find interesting cases
baseline_preds = pd.read_csv('../results/baseline_predictions.csv')

# Get some baseline false negatives (where baseline missed but DistilBERT might catch)
baseline_fn = baseline_preds[
    (baseline_preds['toxic_pred'] == 0) & 
    (baseline_preds['toxic'] == 1)
].head(10)

print("\nChecking if DistilBERT catches baseline false negatives...\n")

for idx in baseline_fn.index[:5]:
    text = baseline_preds.loc[idx, 'comment_text']
    true_labels = [col for col in label_cols if baseline_preds.loc[idx, col] == 1]
    
    # Get DistilBERT prediction
    encoding = tokenizer.encode_plus(
        text,
        max_length=CONFIG['max_length'],
        padding='max_length',
        truncation=True,
        return_tensors='pt'
    )
    
    with torch.no_grad():
        outputs = model(
            input_ids=encoding['input_ids'].to(device),
            attention_mask=encoding['attention_mask'].to(device)
        )
        probs = torch.sigmoid(outputs.logits).cpu().numpy()[0]
        preds = (probs > 0.5).astype(int)
    
    distilbert_labels = [label_cols[i] for i, p in enumerate(preds) if p == 1]
    
    print(f"Text: {text[:150]}...")
    print(f"True labels: {true_labels}")
    print(f"Baseline: Missed (predicted clean)")
    print(f"DistilBERT: {distilbert_labels if distilbert_labels else 'Missed (predicted clean)'}")
    print(f"Toxic probability: {probs[0]:.3f}")
    print("-" * 80)
```
