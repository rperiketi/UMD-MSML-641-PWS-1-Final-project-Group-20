# Toxicity Detection - Error Analysis

# **Goal:** Understand where and why our models fail

### **Key Questions:**
### 1. What types of comments are misclassified?
### 2. Are there linguistic patterns in errors?
### 3. What's the difference between baseline and DistilBERT errors?
### 4. Can we categorize error types?


```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re
import warnings
warnings.filterwarnings('ignore')

print(" Imports complete")
```

### 1. Load Data and Predictions

```python
print("Loading predictions...")

# Load baseline predictions
baseline_preds = pd.read_csv('../results/baseline_predictions.csv')

# Load validation data
val_data = pd.read_csv('../data/train.csv')
from sklearn.model_selection import train_test_split
_, val_data = train_test_split(val_data, test_size=0.2, stratify=val_data['toxic'], random_state=42)
val_data = val_data.reset_index(drop=True)

# Load DistilBERT model for predictions
import torch
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification

device = torch.device('cuda' if torch.cuda.is_available() 
                     else 'mps' if torch.backends.mps.is_available() 
                     else 'cpu')

print(f"Loading DistilBERT model on {device}...")
tokenizer = DistilBertTokenizer.from_pretrained('../models/distilbert')
distilbert_model = DistilBertForSequenceClassification.from_pretrained('../models/distilbert')
distilbert_model = distilbert_model.to(device)
distilbert_model.eval()

print(" Models loaded")

label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']


# Get DistilBERT predictions for validation set
print("Getting DistilBERT predictions...")

def get_distilbert_prediction(text):
    encoding = tokenizer.encode_plus(
        text,
        max_length=128,
        padding='max_length',
        truncation=True,
        return_tensors='pt'
    )
    
    with torch.no_grad():
        outputs = distilbert_model(
            input_ids=encoding['input_ids'].to(device),
            attention_mask=encoding['attention_mask'].to(device)
        )
        probs = torch.sigmoid(outputs.logits).cpu().numpy()[0]
        preds = (probs > 0.5).astype(int)
    
    return preds, probs

# Get predictions for a sample (full set takes too long)
sample_size = 5000
sample_indices = np.random.RandomState(42).choice(len(val_data), sample_size, replace=False)
val_sample = val_data.iloc[sample_indices].copy().reset_index(drop=True)

distilbert_preds = []
distilbert_probs = []

from tqdm.auto import tqdm
for idx in tqdm(range(len(val_sample)), desc="Getting DistilBERT predictions"):
    text = val_sample.iloc[idx]['comment_text']
    preds, probs = get_distilbert_prediction(text)
    distilbert_preds.append(preds)
    distilbert_probs.append(probs)

distilbert_preds = np.array(distilbert_preds)
distilbert_probs = np.array(distilbert_probs)

print(f" Got predictions for {len(val_sample):,} samples")

# %%
# Add predictions to dataframe
for i, label in enumerate(label_cols):
    val_sample[f'distilbert_{label}_pred'] = distilbert_preds[:, i]
    val_sample[f'distilbert_{label}_proba'] = distilbert_probs[:, i]

print(" Predictions added to dataframe")
```

### 2. Error Type Classification

```python
print("\n" + "="*60)
print("ERROR TYPE ANALYSIS")
print("="*60)

# Focus on main 'toxic' label
val_sample['toxic_true'] = val_sample['toxic']
val_sample['toxic_baseline_pred'] = (val_sample['comment_text'].isin(
    baseline_preds[baseline_preds['toxic_pred'] == 1]['comment_text']
)).astype(int)

# Classify errors
def classify_error(row):
    true_label = row['toxic_true']
    baseline_pred = row['toxic_baseline_pred']
    distilbert_pred = row['distilbert_toxic_pred']
    
    if true_label == 1 and baseline_pred == 0 and distilbert_pred == 0:
        return "Both_Miss"
    elif true_label == 1 and baseline_pred == 0 and distilbert_pred == 1:
        return "BERT_Catches"
    elif true_label == 1 and baseline_pred == 1 and distilbert_pred == 0:
        return "Baseline_Catches"
    elif true_label == 0 and baseline_pred == 1 and distilbert_pred == 1:
        return "Both_FP"
    elif true_label == 0 and baseline_pred == 1 and distilbert_pred == 0:
        return "Baseline_FP"
    elif true_label == 0 and baseline_pred == 0 and distilbert_pred == 1:
        return "BERT_FP"
    elif true_label == 1 and baseline_pred == 1 and distilbert_pred == 1:
        return "Both_Correct"
    else:
        return "Both_Correct_Negative"

val_sample['error_type'] = val_sample.apply(classify_error, axis=1)

# %%
# Count error types
error_counts = val_sample['error_type'].value_counts()

print("\nError Type Distribution:")
print(f"{'Error Type':<30} {'Count':>8} {'Percentage':>12}")
print("-" * 55)
for error_type, count in error_counts.items():
    pct = count / len(val_sample) * 100
    print(f"{error_type:<30} {count:>8,} {pct:>11.1f}%")
```

```python
# Visualize error types
fig, ax = plt.subplots(figsize=(12, 6))

colors = {
    'Both_Correct': 'green',
    'Both_Correct_Negative': 'lightgreen',
    'BERT_Catches': 'blue',
    'Baseline_Catches': 'orange',
    'Both_Miss': 'red',
    'Both_FP': 'darkred',
    'BERT_FP': 'pink',
    'Baseline_FP': 'lightsalmon'
}

error_counts_sorted = error_counts.sort_values(ascending=True)
bars = ax.barh(error_counts_sorted.index, error_counts_sorted.values,
               color=[colors.get(x, 'gray') for x in error_counts_sorted.index],
               edgecolor='black', linewidth=1.2)

ax.set_xlabel('Count', fontsize=12)
ax.set_title('Error Type Distribution', fontsize=14, fontweight='bold')
ax.grid(axis='x', alpha=0.3)

# Add value labels
for i, (idx, val) in enumerate(error_counts_sorted.items()):
    ax.text(val + 20, i, f'{val:,}', va='center', fontsize=10)

plt.tight_layout()
plt.savefig('../results/error_type_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Saved: results/error_type_distribution.png")
```

### 3. False Positives Analysis

```python
print("\n" + "="*60)
print("FALSE POSITIVES - Where Models Overpredict")
print("="*60)

# Both models predict toxic, but actually clean
both_fp = val_sample[val_sample['error_type'] == 'Both_FP'].copy()

print(f"\nTotal 'Both False Positives': {len(both_fp):,}")
print("\nSample false positives (both models got wrong):\n")

for idx in range(min(10, len(both_fp))):
    row = both_fp.iloc[idx]
    print(f"[{idx+1}] Text: {row['comment_text'][:200]}...")
    print(f"    True label: Clean")
    print(f"    Baseline confidence: {row.get('toxic_proba', 'N/A')}")
    print(f"    DistilBERT confidence: {row['distilbert_toxic_proba']:.3f}")
    print(f"    Actual labels: {[col for col in label_cols if row[col] == 1]}")
    print("-" * 80)

# %%
# Analyze patterns in false positives
print("\n--- COMMON PATTERNS IN FALSE POSITIVES ---")

def analyze_text_patterns(texts):
    """Analyze common patterns in texts"""
    patterns = {
        'has_profanity': 0,
        'all_caps_words': 0,
        'has_exclamation': 0,
        'has_you': 0,
        'length_long': 0,
        'has_insult_words': 0
    }
    
    profanity = ['fuck', 'shit', 'damn', 'ass', 'hell', 'crap', 'stupid', 'idiot', 'dumb']
    insult_words = ['moron', 'loser', 'jerk', 'fool', 'dick']
    
    for text in texts:
        text_lower = text.lower()
        
        if any(word in text_lower for word in profanity):
            patterns['has_profanity'] += 1
        if any(word.isupper() and len(word) > 2 for word in text.split()):
            patterns['all_caps_words'] += 1
        if '!' in text:
            patterns['has_exclamation'] += 1
        if ' you ' in text_lower or text_lower.startswith('you '):
            patterns['has_you'] += 1
        if len(text) > 200:
            patterns['length_long'] += 1
        if any(word in text_lower for word in insult_words):
            patterns['has_insult_words'] += 1
    
    return patterns

both_fp_patterns = analyze_text_patterns(both_fp['comment_text'].tolist())

print("\nPatterns in 'Both False Positives':")
for pattern, count in both_fp_patterns.items():
    pct = count / len(both_fp) * 100
    print(f"  {pattern:20s}: {count:5,} ({pct:5.1f}%)")

# %%
# Compare with true positives
true_toxic = val_sample[val_sample['toxic_true'] == 1]
true_toxic_patterns = analyze_text_patterns(true_toxic['comment_text'].tolist())

print("\nPatterns in True Toxic Comments (for comparison):")
for pattern, count in true_toxic_patterns.items():
    pct = count / len(true_toxic) * 100
    print(f"  {pattern:20s}: {count:5,} ({pct:5.1f}%)")

print("\n💡 Insight: False positives often contain profanity or strong language")
print("   but in non-toxic contexts (e.g., 'This is fucking awesome!')")
```

### 4. False Negatives Analysis

```python
print("\n" + "="*60)
print("FALSE NEGATIVES - Where Models Miss Toxicity")
print("="*60)

# Both models miss
both_miss = val_sample[val_sample['error_type'] == 'Both_Miss'].copy()

print(f"\nTotal 'Both Miss': {len(both_miss):,}")
print("\nSample false negatives (both models missed):\n")

for idx in range(min(10, len(both_miss))):
    row = both_miss.iloc[idx]
    print(f"[{idx+1}] Text: {row['comment_text'][:200]}...")
    print(f"    True label: TOXIC")
    print(f"    DistilBERT confidence: {row['distilbert_toxic_proba']:.3f}")
    print(f"    Other labels: {[col for col in label_cols if row[col] == 1]}")
    print("-" * 80)

# %%
# Categorize false negatives
print("\n--- CATEGORIZING FALSE NEGATIVES ---")

def categorize_false_negative(text):
    """Categorize why a toxic comment was missed"""
    text_lower = text.lower()
    categories = []
    
    # Check for obfuscation
    if any(char in text for char in ['$', '@', '*', '!']):
        # Check if likely obfuscated profanity
        obfuscated = ['$h!t', 'fu*k', 'a$$', 'b!tch', '@ss', 'f***']
        if any(word in text_lower for word in obfuscated):
            categories.append('obfuscated')
    
    # Check for subtle insults
    subtle_insults = ['pathetic', 'worthless', 'pointless', 'waste', 'useless']
    if any(word in text_lower for word in subtle_insults):
        categories.append('subtle_insult')
    
    # Check for implicit threat
    threat_words = ['better watch', 'you will regret', 'be sorry', 'should be']
    if any(phrase in text_lower for phrase in threat_words):
        categories.append('implicit_threat')
    
    # Check if very short
    if len(text.split()) < 5:
        categories.append('very_short')
    
    # Check if all caps
    if text.isupper() and len(text) > 10:
        categories.append('all_caps')
    
    # Check for identity attacks without slurs
    identity_phrases = ['people like you', 'your kind', 'you people']
    if any(phrase in text_lower for phrase in identity_phrases):
        categories.append('identity_attack')
    
    if not categories:
        categories.append('other')
    
    return categories

both_miss['fn_categories'] = both_miss['comment_text'].apply(categorize_false_negative)

# Flatten categories
all_fn_categories = []
for cats in both_miss['fn_categories']:
    all_fn_categories.extend(cats)

fn_category_counts = Counter(all_fn_categories)

print("\nFalse Negative Categories:")
for category, count in fn_category_counts.most_common():
    pct = count / len(both_miss) * 100
    print(f"  {category:20s}: {count:5,} ({pct:5.1f}%)")

# %%
# Visualize FN categories
fig, ax = plt.subplots(figsize=(10, 6))

categories = list(fn_category_counts.keys())
counts = list(fn_category_counts.values())

bars = ax.barh(categories, counts, color='coral', edgecolor='black', linewidth=1.2)
ax.set_xlabel('Count', fontsize=12)
ax.set_title('False Negative Categories (Both Models Miss)', fontsize=14, fontweight='bold')
ax.grid(axis='x', alpha=0.3)

for i, (cat, val) in enumerate(zip(categories, counts)):
    ax.text(val + 1, i, f'{val}', va='center', fontsize=10)

plt.tight_layout()
plt.savefig('../results/false_negative_categories.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Saved: results/false_negative_categories.png")
```

### 5. DistilBERT Improvements

```python
print("\n" + "="*60)
print("WHERE DISTILBERT IMPROVES OVER BASELINE")
print("="*60)

# Cases where DistilBERT catches but baseline misses
bert_catches = val_sample[val_sample['error_type'] == 'BERT_Catches'].copy()

print(f"\nTotal 'DistilBERT Catches': {len(bert_catches):,}")
print("(Baseline missed, but DistilBERT correctly identified as toxic)\n")

print("Sample cases:\n")
for idx in range(min(15, len(bert_catches))):
    row = bert_catches.iloc[idx]
    print(f"[{idx+1}] Text: {row['comment_text'][:200]}...")
    print(f"    DistilBERT confidence: {row['distilbert_toxic_proba']:.3f}")
    print(f"    True labels: {[col for col in label_cols if row[col] == 1]}")
    print("-" * 80)

# %%
# Analyze what makes these different
bert_catches_patterns = analyze_text_patterns(bert_catches['comment_text'].tolist())

print("\n--- PATTERNS IN BERT CATCHES ---")
for pattern, count in bert_catches_patterns.items():
    pct = count / len(bert_catches) * 100
    print(f"  {pattern:20s}: {count:5,} ({pct:5.1f}%)")

print("\n💡 Insight: DistilBERT better at:")
print("   - Understanding context and word order")
print("   - Detecting implicit toxicity without explicit profanity")
print("   - Handling obfuscated text")
```

### 6. Text Length Analysis


```python
print("\n" + "="*60)
print("TEXT LENGTH vs ERROR PATTERNS")
print("="*60)

val_sample['text_length'] = val_sample['comment_text'].str.len()
val_sample['word_count'] = val_sample['comment_text'].str.split().str.len()

# Analyze by error type
length_by_error = val_sample.groupby('error_type').agg({
    'text_length': ['mean', 'median'],
    'word_count': ['mean', 'median']
})

print("\nText statistics by error type:")
print(length_by_error)

# %%
# Visualize
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

error_types_of_interest = ['Both_Correct', 'Both_Miss', 'BERT_Catches', 'Both_FP']
filtered = val_sample[val_sample['error_type'].isin(error_types_of_interest)]

# Text length
filtered.boxplot(column='text_length', by='error_type', ax=axes[0])
axes[0].set_title('Text Length by Error Type', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Error Type', fontsize=11)
axes[0].set_ylabel('Character Count', fontsize=11)
axes[0].get_figure().suptitle('')

# Word count
filtered.boxplot(column='word_count', by='error_type', ax=axes[1])
axes[1].set_title('Word Count by Error Type', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Error Type', fontsize=11)
axes[1].set_ylabel('Word Count', fontsize=11)
axes[1].get_figure().suptitle('')

plt.tight_layout()
plt.savefig('../results/text_length_by_error.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Saved: results/text_length_by_error.png")
```
